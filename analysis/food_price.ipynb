```python
import pandas as pd

#load the csv file
data = pd.read_csv("../data/wfp_food_prices_ken.csv")

#removes unwanted columns
data = data.drop(columns=['longitude','latitude','admin1','admin2','category','priceflag','usdprice'])

#removes the row on index 0
data = data.drop(index=0)

data = data[
     #filters the market column to Nairobi only
    (data['market'] == 'Nairobi') &
    #filters the unit column to KG only
    (data['unit'] == 'KG') &
    #filter the commodity column to maize
    (data['commodity'] == 'Maize') &
    #filter the pricetype column to wholesale only
    (data['pricetype'] == 'Wholesale')
]

data['price'] = pd.to_numeric(data['price'], errors='coerce')
data['date'] = pd.to_datetime(data['date'], errors='coerce')

data = data.reset_index(drop=True)
data = data.dropna(subset=['price','date'])
#displays the first 25 rows of the filtered data
print(data)

data.to_csv("wfp_food_prices_ken_cleaned.csv", index=False)
```
